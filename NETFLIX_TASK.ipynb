```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("netflix_titles.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset["date_added"] = dataset["date_added"].str.strip(" ")
```

```python
dataset["date_added"] = pd.to_datetime(dataset["date_added"] , format  = "%B %d, %Y")
```

```python
dataset.info()
```

```python
dataset
```

```python
dataset["type"].unique()
```

```python
movies = dataset[dataset["type"] == "Movie"]
TV_shows = dataset[dataset["type"] == "TV Show"]
```

# Frequency of Movies VS TV_Series

```python
plt.figure(figsize = (6,4))
plt.bar(dataset["type"].value_counts().keys(),height = dataset["type"].value_counts(),color = "Silver")
plt.xlabel("type", fontweight="bold", size = 14,color = "Black")
plt.ylabel("count", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)

plt.show()
```

# Movies and TV Series Ratio in UK

```python
plt.figure(figsize = (6,4))
plt.bar(dataset[dataset["country"]=="United Kingdom"]["type"].value_counts().keys(),height = dataset[dataset["country"]=="United Kingdom"]["type"].value_counts(),color = "Silver")
plt.xlabel("type", fontweight="bold", size = 14,color = "Black")
plt.ylabel("count", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)
plt.show()
```

# Top 10 movie producing countries

```python
plt.figure(figsize = (14,6))
plt.bar(movies["country"].value_counts().keys()[:10],height = movies["country"].value_counts()[:10],color = "Silver",width = 0.3)
plt.xlabel("Countries", fontweight="bold", size = 14,color = "Black")
plt.ylabel("# of Movies", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)

plt.show()
```

# Top 10 Tv_Series Producing Countries

```python
plt.figure(figsize = (16,6))
plt.barh(TV_shows["country"].value_counts().keys()[:10] ,TV_shows["country"].value_counts()[:10],color = "Silver")
plt.xlabel("Countries", fontweight="bold", size = 14,color = "Black")
plt.ylabel("# of TV_Series", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)

plt.show()
```

```python
plt.figure(figsize = (14,8))
plt.bar(movies["release_year"].value_counts().keys()[:10],height = movies["release_year"].value_counts()[:10],color = "Silver")
plt.xlabel("Release Year", fontweight="bold", size = 14,color = "BLack")
plt.ylabel("# of Movies", fontweight="bold",size = 14, color = "Black")
plt.xticks([])
plt.tick_params(bottom=False , left = False)

plt.show()
```

```python
plt.figure(figsize = (14,8))
plt.barh(TV_shows["release_year"].value_counts().keys(),TV_shows["release_year"].value_counts(),color = "Silver",height = 0.8)
plt.xlabel("Release Year", fontweight="bold", size = 14,color = "Black")
plt.ylabel("# of Movies", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)

plt.show()
```

```python
plt.figure(figsize = (14,8))
plt.bar(movies["rating"].value_counts().keys(),height = movies["rating"].value_counts(),color = "Silver")
plt.xlabel("Rating", fontweight="bold", size = 14,color = "Black")
plt.ylabel("# of Movies", fontweight="bold",size = 14, color = "Black")
plt.tick_params(bottom=False , left = False)

plt.show()
```
